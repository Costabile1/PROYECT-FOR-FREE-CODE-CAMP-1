```python
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error
```

```python
y_real = df['DatosReales']
y_pred = df['Predicciones']
mse = mean_squared_error(y_real, y_pred)
print("MSE:", mse)

# Calcular el coeficiente de determinación R^2
r2 = r2_score(y_real, y_pred)
print("R^2 Score:", r2)
```

```python
num_predicciones_correctas = (df['Predicciones'] == df['DatosReales']).sum()

# Calcula el número total de predicciones
num_total_predicciones = len(df)

# Calcula la precisión
precision = num_predicciones_correctas / num_total_predicciones

print("Precisión:", precision)
```
